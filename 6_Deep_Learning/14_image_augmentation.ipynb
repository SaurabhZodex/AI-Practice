```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('./Datasets/cat.jpg', target_size=(200,200))
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,          # Randomly rotates images within a range of 40 degrees.
    width_shift_range=0.2,      # Randomly shifts images horizontally (20% of total width).
    height_shift_range=0.2,     # Randomly shifts images vertically (20% of total height).
    # rescale=1./255,             # Rescales pixel values to the range [0, 1] (normalization).
    brightness_range=(0.1, 1.0),# Adjusts brightness from very dark (0.1) to normal brightness (1.0).
    channel_shift_range=0.0,      # No channel shifting, preserving original RGB intensity.
    shear_range=0.2,            # Applies random shearing transformations.
    zoom_range=0.2,             # Randomly zooms in or out by 20%.
    horizontal_flip=True,       # Randomly flips images horizontally.
    fill_mode='nearest'         # Fills any missing pixels after transformation with nearest pixel values.
)
```

```python
img = image.img_to_array(img)
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 200, 200, 3)
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='./Datasets/Augmented_Img'):
    i += 1
    if i == 10:
        break
```
